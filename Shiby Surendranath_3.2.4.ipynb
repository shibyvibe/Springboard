```python
def getSeeClickFix(page=1, pulled=0, search_params={'place_url':'district-of-columbia', 'after': '2020-10-01T00:00:00Z', 'per_page':100}):
    base_url = 'https://seeclickfix.com/api/v2/issues'
    search_params["page"] = page
    seeClickResponse = requests.get(url=base_url, params=search_params)
    if seeClickResponse.ok:
        return seeClickResponse.json()
    else:
        errMsg = (f"Exception: {0}","Failed")
        raise errMsg
        
```

```python
nextPage = 1
while nextPage != None:
    data = getSeeClickFix(page=nextPage)
    print(f"Current Page: {nextPage}")
    #write the data to a file
    with open(f"seeclickfix{nextPage}.json", "a") as f:
        json.dump(data, f)
    nextPage = data["metadata"]["pagination"]["next_page"]

```
